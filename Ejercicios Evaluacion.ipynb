# **Jacobi**

```
import numpy as np
import matplotlib.pyplot as plt
```

```
print('Introduce el valor inicial:')
valor = int(input())
ite = int(input("Introduce el numero de iteraciones:"))
tolera=0.001
contador = 0

x=valor
y=valor
z=valor
tabla=[]

while (contador < ite):
 fx=-0.222*(y) + 0.111*(z)-0.222
 fy=(-0.875*(x) - 0.625*(z)+0.375)
 fz= 0.3*(x)+0.4*(y)-0.6
 tabla.append([ fx,fy,fz ] )
 x=fx
 y=fy
 z=fz
 contador+=1
print(['x', 'y', 'z'])
tabla=np.array(tabla)
print(tabla)
plt.plot(tabla)
plt.title("Jacobi ")
plt.grid()
plt.show()
```

# **Gauss-Siedel**

```
#Gauss_Seidel
print('Introduce el valor inicial:')
valor = int(input())
ite = int(input("Introduce el numero de iteraciones:"))
tolera=0.001
contador = 0

x=valor
y=valor
z=valor
tablax1=[]
tablax2=[]
tablax3=[]

while (contador < ite):
 fx=-0.222*(y) + 0.111*(z)-0.222
 fy=(-0.875*(fx) - 0.625*(z)+0.375)
 fz= 0.3*(fx)+0.4*(fy)-0.6
 tablax1.append([ fx ] )
 tablax2.append([ fy ] )
 tablax3.append([ fz ] )
 x=fx
 y=fy
 z=fz
 contador+=1

print(['x', str(tablax1)])
print(['y', str(tablax2)])
print(['x', str(tablax3)])

print()
plt.plot(tablax1)
plt.plot(tablax2)
plt.plot(tablax3)

plt.title("Gauss_Seidel ")
plt.grid()
plt.show()
```
